```python
#importing required libraries
import pandas as pd
import re
```

```python
# Reading contracts sentences
df=pd.read_csv("Results/Manual-Eval-ChatGPT-final.csv")
df['Individual Scores'][62]
df
```

```python
def Usefulness(scores):
    num=[int(s) for s in re.findall(r'\d+',scores)]
    temp=re.findall(r'[a-zA-Z ]+',scores)
    metrics=['Grammar','Relevant','Logical','Specificity']
    no_sentences =len(num)/len(metrics)
    no_sentences= int(no_sentences)
    # print(no_sentences)
    # print(num)
    d={}
    for i in range(no_sentences):
        d[i]=0

    for i in range(len(num)-no_sentences):
        if num[i]==0 and d[i%no_sentences]!=1:
            d[i%no_sentences]=1
    j=0
    for i in range(len(num)-no_sentences,len(num)):
        if num[i]==0 or num[i]==1 or num[i]==2:
            d[j]=1
        j+=1
    factor=0
    usefulness=0
    for i in range(len(d)):
        if d[i]==1:
            factor+=1
    usefulness= no_sentences-factor
    return "Usefulness: "+ str(usefulness)
```

```python
for i in range(len(df)):
    print(i)
    if df['Final Clarification Questions'][i]!="No Clarification Needed":
        df['Group Scores'][i]=Usefulness(df['Individual Scores'][i])+"\n"+df['Group Scores'][i].lstrip()
    else:
        df['Group Scores'][i]= "Usefulness: 0"+"\n"+df['Group Scores'][i].lstrip()

df.to_csv("Results/Usefulness/Model1-Manual-Evaluation-ChatGPT_final-completed_usefulness.csv",index=False)
```
