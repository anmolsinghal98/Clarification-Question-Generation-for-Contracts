```python
#importing required libraries
import pandas as pd
import re
```

```python
# Reading contracts sentences
df=pd.read_csv("Results/Usefulness/Model1-Manual-Evaluation-ChatGPT_final-completed_usefulness.csv")
df['avg_eval']=""
df
```

```python
Grammar=[]
Relevant=[]
# Clear=[]
Logical=[]
# Neutral=[]
# Seeking_New_Info=[]
Specificity=[]
Sentences=[]
```

```python
def evaluate_individual(a):
    # print(a)
    # print(re.findall(r'\d+', a))
    # sent=len(a.split("\n"))Sentences
    # print(a.split("\n"))
    num=[int(s) for s in re.findall(r'\d+', a)]
    # print(num)
    temp=re.findall(r'[a-zA-Z ]+',a)
    metrics=['Grammar','Relevant','Logical','Specificity']
    # print(metrics)
    no_sentences =len(num)/len(metrics)
    no_sentences= int(no_sentences)
    # print(no_sentences)
    c=0
    nw_metrics=[]
    for i in range(len(metrics)):
        temp=num[c:c+no_sentences]
        c+=no_sentences
        avg=sum(temp)/no_sentences
        nw_metrics.append(metrics[i]+": "+ str(round(avg, 2)))
        if metrics[i] == "Grammar":
            Grammar.append(round(avg, 2))
        elif metrics[i] == "Relevant":
            Relevant.append(round(avg, 2))
#         elif metrics[i] == "Clear":
#             Clear.append(round(avg, 2))
        elif metrics[i] == "Logical":
            Logical.append(round(avg, 2))
#         elif metrics[i] == "Neutral":
#             Neutral.append(round(avg, 2))
#         elif metrics[i] == "Seeking New Info":
#             Seeking_New_Info.append(round(avg, 2))
        elif metrics[i] == "Specificity":
            Specificity.append(round(avg, 2))
    Sentences.append(no_sentences)
    return '\n'.join(nw_metrics)
```

```python
count=0
for i in range(len(df)):
    if df['Final Clarification Questions'][i]!="No Clarification Needed":
        df['avg_eval'][i]=evaluate_individual(df['Individual Scores'][i])
        # print(i)
    else:
        count+=1
deno=len(df)-count
df.to_csv("Results/Final_results/Model1-Amrita-completed_eval.csv",index=False)
# Sentences
```

```python
print("Grammar: " + str(sum(Grammar)/deno))
print("Relevant: " + str(sum(Relevant)/deno))
# print("Clear: " + str(sum(Clear)/deno))
print("Logical: " + str(sum(Logical)/deno))
# print("Neutral: " + str(sum(Neutral)/deno))
# print("Seeking New Info: " + str(sum(Seeking_New_Info)/deno))
print("Specificity: " + str(sum(Specificity)/(4*deno)))
```

```python
redundant=0
usefulness=0
Missing_Cases=[]
```

```python
def evaluate_group(b):
    b=b.split("\n")
    # print(b)
    use=re.findall(r'\d+', b[0])
    # print(b)
    useful=int(use[0])
    redu=b[1].split("AND")
    # print(redu)
    misca=re.findall(r'\d+', b[2])
    # print(misca)
    # print(useful)
    # print(misca)
    Missing_Cases.append(misca[0])
    Redundant=0
    for j in redu:
        # print(redu)
        # print(len(re.findall(r'\d+', j)))
        Redundant+=len(re.findall(r'\d+', j))-1
        if Redundant==-1:
            Redundant=0
    return useful,Redundant
```

```python
for i in range(len(df)):
    # print(i)
    u,r=evaluate_group(df['Group Scores'][i].lstrip())
    usefulness+=u
    redundant+=r
```

```python
print("Total Questions: ",sum(Sentences))
print("Total Useful: ",usefulness)
print("Total Reduandant: ",redundant)
print("Count of 0:", Missing_Cases.count('0'))
print("Count of 1:", Missing_Cases.count('1'))
print("Count of 2:", Missing_Cases.count('2'))
```
